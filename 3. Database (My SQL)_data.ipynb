# Imporitng Data from My SQL Database

## Using mysql.connector 

```python
import mysql.connector

# making connection
conn = mysql.connector.connect(host = 'localhost', user = 'root', password = '', database = 'world')

# query
query = 'SELECT * FROM city'
pd.read_sql_query(query, conn)
```

### Note : pandas use only sqlalchemy connction . It will throw error when we try to read sql table using pd.read_sql_table using mysql.connector

## Using sqlalchemy connection

```python
from sqlalchemy import create_engine

username = 'root'
password = ''
host = 'localhost'
database = 'world'

engine = create_engine(f"mysql+mysqlconnector://{username}:{password}@{host}/{database}")

pd.read_sql_table('city', con = engine)
```
