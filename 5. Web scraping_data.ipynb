# Web Scraping Using BeautifulSoup

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# scraping data from Nav page
url = 'https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav'
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
response = requests.get(url, headers = headers)
soup = BeautifulSoup(response.text)

company_name = []
rating = []
reviews = []
salaries = []
interviews = []
jobs = []
about = []


containers = soup.find_all('div', class_ = 'companyCardWrapper')

for container in containers:  
        company_name.append(container.find('h2').text.strip())
        rating.append(container.find('div', class_ = 'rating_text').text.strip())
        reviews.append(container.find_all('a', class_ = 'companyCardWrapper__ActionWrapper')[0].text.strip())
        salaries.append(container.find_all('a', class_ = 'companyCardWrapper__ActionWrapper')[1].text.strip())
        interviews.append(container.find_all('a', class_ = 'companyCardWrapper__ActionWrapper')[2].text.strip())
        jobs.append(container.find_all('a', class_ = 'companyCardWrapper__ActionWrapper')[3].text.strip())
        about.append(container.find('span', class_ = 'companyCardWrapper__interLinking').text.strip())
    
    
    
final_df = pd.DataFrame({
        'company' : company_name,
        'rating' : rating, 
        'review' : reviews,
        'interview' : interviews,
        'salary' : salaries,
        'jobs' : jobs,
        'about': about
    })
final_df
```

```python
# Scraping data from rest of the pages
for page in range(2, 20):
    company_name = []
    rating = []
    reviews = []
    salaries = []
    interviews = []
    jobs = []
    about = []
    
    url = f'https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page={page}'
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    response = requests.get(url, headers = headers)
    soup = BeautifulSoup(response.text)

    containers = soup.find_all('div', class_ = 'companyCardWrapper')

    for container in containers:  
        company_name.append(container.find('h2').text.strip())
        rating.append(container.find('div', class_ = 'rating_text').text.strip())
        reviews.append(container.find_all('a', class_ = 'companyCardWrapper__ActionWrapper')[0].text.strip())
        salaries.append(container.find_all('a', class_ = 'companyCardWrapper__ActionWrapper')[1].text.strip())
        interviews.append(container.find_all('a', class_ = 'companyCardWrapper__ActionWrapper')[2].text.strip())
        jobs.append(container.find_all('a', class_ = 'companyCardWrapper__ActionWrapper')[3].text.strip())
        about.append(container.find('span', class_ = 'companyCardWrapper__interLinking').text.strip())
    
    
    
    df = pd.DataFrame({
        'company' : company_name,
        'rating' : rating, 
        'review' : reviews,
        'interview' : interviews,
        'salary' : salaries,
        'jobs' : jobs,
        'about': about
    })
    df

    final_df = pd.concat((final_df, df))


final_df
    
```
